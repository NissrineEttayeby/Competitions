```python
#define x , y train and x test

X = df1.iloc[::,1:]
y = df1["label"]
X_test = df2
```

# Scaling the dataset

```python
# the dataset has images with pixeles values in the range [0,255]
X = X / 255
X_test = X_test / 255
```

```python
X_test
```

```python
X.shape
```

```python
ImageId = X_test.index
ImageId 
```

```python
from sklearn.model_selection import train_test_split
X_train , X_val , y_train , y_val = train_test_split(X,y, test_size=0.1, random_state=1)
X_train.shape , X_val.shape, y_train.shape , y_val.shape , X_test.shape
```

```python
X_val
```

```python
y_val
```

# Model ML

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(hidden_layer_sizes=(100,), activation='relu', solver='adam', alpha=0.0001, batch_size='auto', learning_rate='constant', learning_rate_init=0.001, power_t=0.5, max_iter=200, shuffle=True, random_state=None, tol=0.0001, verbose=False, warm_start=False, momentum=0.9, nesterovs_momentum=True)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_val)
y_pred
```

```python
print(model.score(X_val, y_val))
```

```python
df_val = pd.DataFrame({'true_value': y_val , 'predicted values': y_pred})
```

```python
df_val.reset_index(drop = True , inplace = True)
```
